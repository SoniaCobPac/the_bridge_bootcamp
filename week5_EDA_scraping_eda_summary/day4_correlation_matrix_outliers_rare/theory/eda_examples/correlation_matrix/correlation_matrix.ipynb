```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# MPG: Millas por galón
# MSRP: Precio de venta

df = pd.read_csv("HP_engine.csv")
df.sample(10)
```

```python
df.shape
```

```python
# De momento solo vamos a analizar columnas numéricas
df.info()
```

```python
# Correlación de Pearson
# -1 < r < 1

# Finding the relations between the variables.
plt.figure(figsize=(15,10))
c = df.corr()   #matriz de correlación
sns.heatmap(c, cmap="BrBG",annot=True)  #estamos haciendo la distribución de valores que tiene el heatmap
c
```
