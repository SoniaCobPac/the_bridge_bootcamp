### Step 1. Import the necessary libraries

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
'% matplotlib inline'

sns.set_style("white")
```

### Step 2. Create the DataFrame it should look like below.

This time you will create the data.

```python
first_name = "Jason", "Molly", "Tina", "Jake", "Amy"
last_name = "Miller", "Jacobson", "Ali", "Milner", "Cooze"
age = [42, 52, 36, 24, 73]
preTestScore = [4, 24, 31, 2, 3]
postTestScore = [25, 94, 57, 62, 70]

df = pd.DataFrame({"first_name":first_name, "last_name": last_name, "age": age, "preTestScore": preTestScore, "postTestScore": postTestScore})
df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
ax = sns.scatterplot(x = df["preTestScore"], y = df["postTestScore"], s=100)

ax.set_xlim(0,35)
ax.set_ylim(20,100)
ax.set_title("preTestScore x postTestScore")

plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
sex = [0, 1, 1, 0, 1]
df["sex"] = sex
```

```python
ax = sns.scatterplot(x = df["preTestScore"], y = df["postTestScore"], s=4.5*df["postTestScore"], c=df["sex"], palette="Paired")


ax.set_title("preTestScore x postTestScore")
ax.set_xlim(0,35)
ax.set_ylim(20,100)

plt.show()
```
