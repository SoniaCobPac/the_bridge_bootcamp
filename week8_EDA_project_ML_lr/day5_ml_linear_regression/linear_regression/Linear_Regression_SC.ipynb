### Ejemplo de particion de los datos

```python
# datos totales
X = [5, 80, 40, 24, 8]
y = [105, 180, 140, 124, 108]

# conjunto de entrenamiento. Es importante que no se mantenga un orden específico
X_train = [40, 24, 80, 5]
y_train = [140, 124, 180, 105]

x = X_train[0]
x = 40

# test - el único valor que no esta en el conjunto de entrenamiento
X_test = [8]
y_test = [108]
```

```python
# como antes no podemos usar vectores, tenemos que usar arrays
X = np.array(X)
X = X.reshape(-1,1)

y = np.array(y)
y = y.reshape(-1,1)
X.shape
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

train_errors, val_errors = [], []   #val_errors = test_errors
def plot_learning_curves(model, X, Y):
    # 80% de Train
    # 20% de Test
    # Partimos los datos:
    X_train, X_val, y_train, y_val = train_test_split(X, Y, test_size=0.05) # función a la que le damos tres argumentos
    # test_size es el 5% de todos los datos, entonces x_train sería el 95%
    # y representa la etiqueta (label) de x
    # nosotros elegimos el porcentaje del test_size, pero por defecto serían el 20% de los datos
    
    for m in range(1, len(X_train)):  # Incremental
        x = X_train[:m] 
        # y es el valor que debería dar cuando le das 'x'
        y = y_train[:m]
        # el model es la regresión lineal
        # fit = entrenar la regresión lineal
        model.fit(x, y)

        y_train_predict = model.predict(x)  # que la linea model nos prediga un valor y a partir del valor x. No es lo que vale         y es lo que el model cree que debería dar
        y_val_predict = model.predict(X_val) # los valores (en forma array) que no han sido entrenados, nosotros sabemos los            que deberían dar:- Están guardados en y_train

        # predice el error del conjunto del entrenamiento:
        # diferencia de lo que debería dar y y lo que ha dado el modelo predictivo al darle x
        train_errors.append(mean_squared_error(y, y_train_predict))

        # predice el error del conjunto de test:
        # y_val_predict es el valor que ha dado al darle x_val
        val_errors.append(mean_squared_error(y_val, y_val_predict))
        
        if m < 4:
            print("##### ITERACIÓN " + str(m) + " #####")
            print("x", x)
            print("y", y)
            print("y_train_predict", y_train_predict)
            print("-----\nX_val\n", X_val)
            print("y_val_predict", y_val_predict)
    
    # curvas de los errores de entrenamiento y errores del test
    plt.plot(train_errors, "r-+", linewidth=2, label="train")
    plt.plot(val_errors, "b-", linewidth=3, label="val")
```

```python
from sklearn.linear_model import LinearRegression

# crea una instancia de la regresión lineal
lin_reg = LinearRegression()    #lin_reg es el modelo que vamos a entrenar
plot_learning_curves(model=lin_reg, X=X, Y=y)
```

```python
# y = a + bX

# b, pendiente
#lin_reg es una variable que creamos antes, es el modelo que vamos a entrenar
b = lin_reg.coef_ # reg.coef es un atributo para sacar la pendiente
b
```

```python
# a, secante
a = lin_reg.intercept_  # la secante. Donde corta en el eje y
a
```

```python
print("|||||||||||ERRORES||||||||||||")
print("train_errors\n\n", train_errors)
print("val_errors\n\n", val_errors)
```

### Pasos para resolver un ejercicio de regresión lineal

1. Cargamos el conjunto de datos
2. Partimos el conjunto en train y test
3. Creamos una variable que represente el modelo (inicializamos)
4. Fit (entrenamos) el modelo con el conjunto train
5. Calculamos el error de Train y de Test
6. ¿Es un buen error? --> si no, repetimos el proceso:

    6.1. Cambiando la semilla. La ordenación de los números puede influir en la predicción
    
    6.2. Eligiendo un porcentaje de test diferente

    6.3. Coger más datos
