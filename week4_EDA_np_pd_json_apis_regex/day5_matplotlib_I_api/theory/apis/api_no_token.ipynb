```python
# No todas las APIs devuelven todos los posibles valores debido a limitación de recursos

url = "https://pokeapi.co/api/v2/pokemon"   # como dice el código 'limit=20' nos están dando una lista de 20. Podemos cambiar el límite
poke_json = requests.get(url).json()
poke_json     # Esta es la primera llamada a la API
            
```

```python
# Esta API tiene las claves 'count', 'next', 'previous', 'results' y podemos usarlo para saber como conseguir más información. Pero no significa que todas tengan la misma forma
```

```python
json_normalize(poke_json)
```

```python
# Podemos acceder a esta información porque estoy accediendo al valor con la clave 'next'
# Offset significa desde que objeto pedimos informacion. Si offset es 20, la siguiente vez que pedimos información nos da desde el elemento número 20. 
poke_json["next"]
```

```python
# Cogemos 52 pokemons desde el elemento 20
# A partir del interrogante son los metadatos, que son el offset y el limit
# Si pokemon tuviese algun tipo de limitación para el número de elementos que podemos descargar podríamos poner un límite muy alto para traer los máximos posibles. 
# Si next es none significa que ya no hay más información que recoger

answer = requests.get("https://pokeapi.co/api/v2/pokemon?offset=20&limit=52").json()
answer
```

```python
df_pokemons = pd.DataFrame(answer)
df_pokemons.head()
```

```python
df2 = pd.DataFrame(answer["results"])   # la url me da toda la información de cada pokemon. Para obtener la información de cada uno tenemos que volver a llamar esas APIs
df2.head()
```

```python
pokemon = [] # para contener todos los resultados de la api
poke_dict = {"name":"url"}

pokemon.extend(poke_json["results"])
pokemon
# Voy a crear una lista de diccionario que tenga dos claves, una el nombre del pokemon y la otra va a ser la url de cada pokemon
```

```python
%%time
pokemon_api_url = "https://pokeapi.co/api/v2/pokemon"
poke_json = requests.get(pokemon_api_url).json()
pokemon = []    # En este caso quiero tener la información en una lista, pero no es obligatorio

# Almacenamos 40 pokemons realizando peticiones hasta que la lista de pokemons tenga 40 o más pokemons. Si no limitamos esto la cantidad de información que podemos obtener sería enorme (eje. twitts en un día- no tendría memoria para almacenarla)
while len(pokemon) < 40:
#while len(pokemon) < 1118:
    pokemon.extend(poke_json["results"])
    next_url = poke_json["next"]    # estoy pidiendo los siguientes 20 pokemons
    poke_json = requests.get(next_url).json()   # Peticiones a las API de cada pokemon
print(len(pokemon))
pokemon
```

```python
# Es mejor pedir más informacion de una vez a la API que estar realizando varias llamadas, que es lo que tarda más tiempo
```

```python
%% time # Esto representa el tiempo que esta celda tarda en ejecutarse
```

```python
%%time  
pokemon_api_url = "https://pokeapi.co/api/v2/pokemon"
poke_json = requests.get(pokemon_api_url).json()
pokemon = []

import time
while poke_json["next"] != None:    # Mientras siga habiendo más urls, mientras siga existiendo información
    pokemon.extend(poke_json["results"])
    next_url = poke_json["next"]
    poke_json = requests.get(next_url).json()
    #print(poke_json["next"])
    #time.sleep(1)
pokemon.extend(poke_json["results"])

print(len(pokemon))
pokemon
```

```python
# Podemos convertir la información de la API a diccionario:

diccionario_pokemons_url = {}
for i, e in enumerate(pokemon):
    nombre_pokemon = pokemon[i]["name"]
    url_pokemon = pokemon[i]["url"]
    diccionario_pokemons_url[nombre_pokemon] = url_pokemon

diccionario_pokemons_url
```

```python
df = pd.DataFrame(diccionario_pokemons_url, index=["url"])
df 
# Tenemos que transponer el data frame para tener 1118 rows y 1 columna
```

```python
# Convierto el diccionario a data frame
df = pd.DataFrame(diccionario_pokemons_url, index=["url"])
df = df.T   #.T es el data frame transpuesta
df
```
