# How to Write Type Annotations

### Python < 3.9

```python
# Type annotation es el concepto de escribir el tipo de las variables que voy a utilizar. Es para informar al lector pero no fuerza que las variables y el retorno sea del tipo que estoy especificando
# En este caso está definiendo que el parámetro xs sea una lista
# flechita float representa lo que va a retornar la función

def total(xs: list) -> float:
    return sum(xs)

l = [2, 4, 6, 8.9]
total(xs=l)
```

```python
# No estoy cumpliendo que xs sea una lista o el restultado sea un float. La función funciona pero es mala praxis porque no estoy siguiendo la recomendación del programador

def total_mala_praxis(xs: list) -> float:
    a = list(xs.keys())
    return str(sum(a))

l = {2:"das", 14:"x"}
total_mala_praxis(xs=l)
```

```python
from typing import List  # note capital L

def total(xs: List[float]) -> float:
    return sum(xs)

l = [2.1, 4.5, 4.6, 0.8]
total(xs=l)
```

```python
def f() -> str:
    return "2"

print(type(f()))
```

```python
def greeting(name: str, s:int) -> float:
    """ Documentación """
    return 'Hello ' + name

greeting("2", "a")
```

### Python => 3.9

```python
# Ya no hace falta importar ninguna librería para hacer lo mismo
tipo_lista_de_floats = list[float]

def scale(scalar: float, vector: tipo_lista_de_floats) -> tipo_lista_de_floats:
    return [scalar * num for num in vector]

# typechecks; a list of floats qualifies as a Vector.
new_vector = scale(2.0, [1.0, -4.2, 5.4])
```
