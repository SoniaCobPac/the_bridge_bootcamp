```python
def is_bull(random_number, user_input):
    """
    This function returns the number of same digits in both parameters without considering their position

    Parameters: Two strings

    """
    bulls = 0
    l = list(random_number)
    for i, v in enumerate(user_input):
        if v in random_number and v != random_number[i] and v in l:
            bulls += 1
            l.remove(v)
    return bulls

is_bull(random_number, user_input)

def is_cow(random_numer, user_input):
    """
    This functions returns the number of same digits that are in the same position in both parameters
    
    Parameters: Two strings
    
    """
    return sum([1 for i, v in enumerate(user_input) if v == random_numer[i]])
```

```python
def is_cow(random_numer, user_input):
    lista_c = ['c' if v == random_number[i] else v for i, v in enumerate(user_input)]
    cows = sum([1 for i, v in enumerate(user_input) if v == random_numer[i]])
    return cows, lista_c

def find_cow_bull(random_numer, user_input):
    cows, posibles = is_cow(random_numer, user_input)
    bulls = 0
    for i, v in enumerate(user_input):
        if v in posibles and posibles[i] == random_number[i]:
            bulls += 1
    return cows, bulls
```

```python
#Group C

# -------- Modules --------
import random
from functools import reduce

# -------- Function to check number of cows/bulls in guess --------
def checker(num, guess):
    num = str(num)
    guess = str(guess)

    cows = 0    # Value and position
    bulls = 0   # Position

    for i, x in enumerate(num):
        if x in guess:
            bulls += 1

            if guess[i] == x:
                bulls -= 1
                cows += 1

    return [bulls, cows]

# -------- To create the random number to guess --------
def random_non_repetitive():
    x = []
    count = 0

    while count < 4:
        nr = random.randint(0, 9)
        if nr not in x:
            x.append(str(nr))
        count += 1

    num = int(reduce(lambda x, y: x + y, x))
    return num

# -------- To play the game --------
# I have to pass it a num. In case I want to have a random number, then I pass the function above
def game(num):
    bulls, cows = [0, 0]

    while cows < 4:
        guess = input('Enter your guess: ')
        if guess == 'STOP':
            break

        bulls, cows = checker(num, guess)
        print('{} cows, {} bulls'.format(cows, bulls))

    print('Your guess: ', guess)
    print('Your num: ', num)

# numero a adivinar
#num = int(input('Enter number'))
game(random_non_repetitive())
```
